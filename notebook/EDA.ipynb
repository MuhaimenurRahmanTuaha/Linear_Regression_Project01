## About Data
This is a fitness tracker product dataset consisting of different products from various brands with their specifications, ratings and reviews for the Indian market. The data has been collected from e-commerce websites namely Flipkart and Amazon using web scraping technique.

Inspiration

This dataset could be used to find answers to some interesting questions like -

* Is there a significant demand for fitness trackers in the Indian market?
* Information on the top 5 brands for fitness bands and smart watches
* Is there a correlation between the prices and product specifications, ratings, etc.
* Different types of fitness trackers, their price segments for different users

This dataset contains 451 samples with 16 attributes. There are some missing values in this dataset.

Check duplicate column

```python
def duplicate_cols(data):
    dup_cols = set()
    for x in range(data.shape[1]):
        base_col = data.iloc[:,x]
        for y in range(x+1,data.shape[1]):
            comp_col = data.iloc[:,y]
            if base_col.equals(comp_col):
                dup_cols.add(data.columns.values[y])
    return dup_cols
duplicate_cols(df)
```

There is no duplicate column

**Insights**
* There is no duplicate row and column

#### Differentiate the categorical and Numerical Columns

```python
categorical_columns = [feature for feature in df.columns if df[feature].dtype == "O"]
numerical_columns = [feature for feature in df.columns if df[feature].dtype != "O"]
```

```python
print("Categorical features : {}".format(categorical_columns))
print("Numerical features : {}".format(numerical_columns))
```

## Visual Analysis

```python
fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(14,10))
for i,j in enumerate(numerical_columns):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.histplot(data=df[j],ax=ax)
```

* None of the columns are normally distributed

```python
fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(14,10))
for i,j in enumerate(numerical_columns):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.kdeplot(data=df[j],ax=ax,fill=True)
```

* Current price, original price and number of ratings are right skewed
* Discount percentage, rating and battery life are left skewed

#### Outliers

```python
fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(14,10))
for i,j in enumerate(numerical_columns):
    row = i//3
    col = i%3
    ax = axes[row,col]
    sns.boxplot(data=df[j],ax=ax)
    ax.set_xlabel(j)
```

* Only battery life has no outliers

```python
for feature in categorical_columns:
    fig, axs = plt.subplots(1, 2, figsize=(15, 7))
    data = df.copy()
    plt.subplot(121)
    data.groupby(feature)['Original Price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Original Price')
    plt.title(feature)
    plt.subplot(122)
    data.groupby(feature)['Current Price'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Current Price')
    plt.title(feature)
    plt.show()
```
