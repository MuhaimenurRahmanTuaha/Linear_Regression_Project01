#### **2.2.Check duplicate column**

```python
def duplicate_cols(data):
    dup_cols = set()
    for x in range(data.shape[1]):
        base_col = data.iloc[:,x]
        for y in range(x+1,data.shape[1]):
            comp_col = data.iloc[:,y]
            if base_col.equals(comp_col):
                dup_cols.add(data.columns.values[y])
    return dup_cols
duplicate_cols(df)
```

There is no duplicate column

**Insights**
* There is no duplicate row and column

```python
from scipy.stats import chi2_contingency
```

```python
df[categorical_columns[0]]
```

```python
for i in range(8):
    j=7-i
    for k in range(j):
        contingency_table = pd.crosstab(df[categorical_columns[i]], df[categorical_columns[k+i+1]])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        if p<0.05:
            print("There is a correlation between {0} and {1} and p-value is : {2}".format(categorical_columns[i],categorical_columns[k+i+1],p))
        else:
            print("There is no correlation between {0} and {1} and p-value is : {2}".format(categorical_columns[i],categorical_columns[k+i+1],p))
```

* N.B. : Touchscreen and bluetooth hava no impact on original price and current price (it's surprising as per consider domain knowledge)

### **7.2.Multivariate Analysis**

#### **7.2.1.Relation between "Original Price", "Current Price", "Discount Percentage" and categorical features**

```python
for feature in categorical_columns:
    
    if feature == "Model Name":
        pass

    else: 
        fig, axs = plt.subplots(1, 3, figsize=(20, 7))
        data = df.copy()

        plt.subplot(131)
        data.groupby(feature)['Original Price'].median().plot.bar()
        plt.xlabel(feature)
        plt.ylabel('Original Price')
        plt.title(f"{feature} vs Original Price")

        plt.subplot(132)
        data.groupby(feature)['Current Price'].median().plot.bar()
        plt.xlabel(feature)
        plt.ylabel('Current Price')
        plt.title(f"{feature} vs Current Price")

        plt.subplot(133)
        data.groupby(feature)['Discount Percentage'].median().plot.bar()
        plt.xlabel(feature)
        plt.ylabel('Discount Percentage')
        plt.title(f"{feature} vs Discount Percentage")
        plt.show()
```
