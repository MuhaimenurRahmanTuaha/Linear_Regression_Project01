#### Check duplicates row and column

Check duplicate row

```python
df.duplicated().sum()
```

There is no duplicate row

Check duplicate column

```python
def duplicate_cols(data):
    dup_cols = set()
    for x in range(data.shape[1]):
        base_col = data.iloc[:,x]
        for y in range(x+1,data.shape[1]):
            comp_col = data.iloc[:,y]
            if base_col.equals(comp_col):
                dup_cols.add(data.columns.values[y])
    return dup_cols
duplicate_cols(df)
```

There is no duplicate column

**Insights**
* There is no duplicate row and column

#### Differentiate the categorical and Numerical Columns

```python
categorical_columns = [feature for feature in df.columns if df[feature].dtype == "O"]
numerical_columns = [feature for feature in df.columns if df[feature].dtype != "O"]
```

```python
print("Categorical features : {}".format(categorical_columns))
print("Numerical features : {}".format(numerical_columns))
```

#### Check the unique values in the columns

```python
# Categorical feature unique no
for feature in categorical_columns:
    lenght_of_unique_values = len(df[feature].unique())
    print("'{0}' feature's no of unique values number : '{1}'".format(feature,lenght_of_unique_values))
```

```python
# Numerical feature unique no
for feature in numerical_columns:
    lenght_of_unique_values = len(df[feature].unique())
    print("'{0}' feature's no of unique values number : '{1}'".format(feature,lenght_of_unique_values))
```
